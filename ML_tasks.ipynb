# Classification Task

### Features Engineering

```python
classification=data.copy()

#outcome variable team1_win as a probability of team1 winning the match
classification.loc[classification["winner"]==classification["team1"],"team1_win"]=1
classification.loc[classification["winner"]!=classification["team1"],"team1_win"]=0

#outcome variable team1_toss_win as a value of team1 winning the toss
classification.loc[classification["toss_winner"]==classification["team1"],"team1_toss_win"]=1
classification.loc[classification["toss_winner"]!=classification["team1"],"team1_toss_win"]=0

#outcome variable team1_bat to depict if team1 bats first
classification["team1_bat"]=0
classification.loc[(classification["team1_toss_win"]==1) & (classification["toss_decision"]=="bat"),"team1_bat"]=1
```

```python
classification
```

### Encoding non-numerical vales

```python
#encoding
encoder=preprocessing.LabelEncoder()
classification["team1"]=encoder.fit_transform(classification["team1"])
classification["team2"]=encoder.fit_transform(classification["team2"])
classification["venue"]=encoder.fit_transform(classification["venue"])
```

```python
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.9:
            column = correlation_matrix.columns[i]
            correlated_features.add(column)
```

# Regression Task

### Features Engineering

```python
regression=classification.copy()
df=regression.filter(['match_id','venue','inning','batting_team','bowling_team','batsman','bowler','over','ball','total_runs'],axis=1)
df
```

```python
dfs=[]
for i in df['match_id'].unique():
  for j in range(1,3):
    d=df[(df['match_id']==i) & (df['inning']==j)]
    s=d['total_runs'].sum()
    d['final_score']=s
    current=0
    d['current_score']=0
    for ind in d.index:
      current+=d['total_runs'][ind]
      d['current_score'][ind]=current
    data2 = d.copy()
    data2 = data2.groupby(by=['over']).sum()['total_runs'].reset_index()
    data3 = pd.merge(left=d, right=data2, on='over', how='right')
    dfs.append(data3)
final=pd.concat(dfs)
final
```

The column 'final_score' is the target variable

The column 'current_score' gives the current score

The column 'total_runs_y' gives the total runs in scored in that over

'total_runs_y' will be renamed to 'total_over_runs'

## Loss Function

```python
def custom_accuracy(y_test,y_pred,thresold=10):
    right = 0
    l = len(y_pred)
    for i in range(0,l):
        if(abs(y_pred[i]-y_test[i]) <= thresold):
            right += 1
    return ((right/l)*100)
```

It calculates the accuracy based on whether the predicted final score is in +-10 range of the the actual final score

```python
print("Accuracy: {:.3f} %".format(custom_accuracy(y_test,y_pred,10)))
```

```python
c_s1=custom_accuracy(y_test,y_pred,10)
mae1=mean_absolute_error(y_test,y_pred)
mse1=mean_squared_error(y_test,y_pred)
rmse1=np.sqrt(mean_squared_error(y_test,y_pred))
msle1=mean_squared_log_error(y_test,y_pred)
rmsle1=np.sqrt(mean_squared_log_error(y_test,y_pred))
```
